```python
def tokenizer_comment(comment):
    from konlpy.tag import Twitter
    twitter = Twitter()
    i = 0
    malist = twitter.pos(comment,stem=True)
    r = []
    # 불용어 추가
    stopwords = ['하다', ',', '들', '이', '..', '.', '것', '다', '이다', '~', '그', '그녀', '저', '...', '"', '~~']
    
    remove_tag = ['Determiner', 'Josa', 'Eomi', 'PreEomi', 'URL', 'Email', 'Hashtag', 'Alpha',
                  'ScreenName', 'KoreanParticle', 'Foreign', 'Cashtag']
    
    try:
        for word in malist:
            # 어미/조사/구두점/ㅋㅋ^^ㅎㅎ/음표살림/Alphabet/부사는 대상에서 제외            
            if not word[1] in remove_tag:
                if not word[0] in r:
                    if not word[0] in stopwords:
                        # 숫자, 특수문자 제거.
                        r.append(word[0])
        return r

    except Exception as e:
        print (e)
```

```python
emotions = []
i = 0
check_tokens = []
for comment in token_data:
    check_token = []
    emotion = []
    happy = 0
    sad = 0
    disgust = 0
    surprised = 0
    fear = 0
    angry = 0
    
    # sad
    if('가슴' in comment) &('아프다' in comment):
        check_token.append('가슴아프다2')
        sad+=1
    if('기운' in comment) &('없다' in comment):
        check_token.append('기운없다2')
        sad+=1
    if('남' in comment) &('부럽다' in comment):
        check_token.append('남부럽다2')
        sad+=1
    if('뒤' in comment) &('숭숭' in comment):
        check_token.append('뒤숭숭2')
        sad+=1
    if('맥' in comment) &('빠지다' in comment):
        check_token.append('맥빠지다2')
        sad+=1
    if('복' in comment) &('받치다' in comment):
        check_token.append('복받치다2')
        sad+=1
    if('뼈' in comment) &('아프다' in comment):
        check_token.append('뼈아프다2')
        sad+=1
    if('뼈' in comment) &('저리다' in comment):
        check_token.append('뼈저리다2')
        sad+=1
    if('애' in comment) &('끓다' in comment):
        check_token.append('애끓다2')
        sad+=1
    if('애' in comment) &('닮다' in comment):
        check_token.append('애닮다2')
        sad+=1
    if('애' in comment) &('타다' in comment):
        check_token.append('애타다2')
        sad+=1
    if('처' in comment) &('연하다' in comment):
        check_token.append('처연하다2')
        sad+=1
    if('속' in comment) &('썩다'in comment):
        check_token.append('속썩다2')
        sad+=1
    if('속' in comment) &('앓이' in comment):
        check_token.append('속앓이2')
        sad+=1
    if('속' in comment) &('타다' in comment):
        check_token.append('속타다2')
        sad+=1
    # disgust
    if('못' in comment) &('밉다' in comment):
        check_token.append('못밉다2')
        disgust+=1
    if('정' in comment) &('떨어지다' in comment):
        check_token.append('정떨어지다2')
        disgust+=1
    if('못' in comment) &('마땅하다' in comment):
        check_token.append('못마땅하다2')
        disgust+=1
    if('시' in comment) &('덥다' in comment):
        check_token.append('시덥다2')
        disgust+=1
    if('남' in comment) &('부끄럽다' in comment):
        check_token.append('남부끄럽다2')
        disgust+=1
    if('징' in comment) &('그렇다' in comment):
        check_token.append('징그렇다2')
        disgust+=1
    # fear
    if ('뒤' in comment) & ('숭숭' in comment):
        check_token.append('뒤숭숭2')
        fear += 1
    # surprised
    if ('기상' in comment) & ('천외' in comment):
        check_token.append('기상천외2')
        surprised += 1
    if ('얼' in comment) & ('떨다' in comment):
        check_token.append('얼떨다2')
        surprised += 1
    # angry    
    if ('못' in comment) & ('되다' in comment):
        check_token.append('못되다2')
        angry += 1    
    if ('삐' in comment) & ('딱하다' in comment):
        check_token.append('삐딱하다2')
        angry += 1   
    if ('비' in comment) & ('딱하다' in comment):
        check_token.append('비딱하다2')
        angry += 1   
    if ('약' in comment) & ('오르다' in comment):
        check_token.append('약오르다2')
        angry += 1    
    if ('치' in comment) & ('떨리다' in comment):
        check_token.append('치떨리다2')
        angry += 1    
    if ('치밀' in comment) & ('오르다' in comment):
        check_token.append('치밀어오르다2')
        angry += 1
    if ('성' in comment) & ('나다' in comment):
        check_token.append('성나다2')
        angry += 1
    if ('못' in comment) & ('마땅하다' in comment):
        check_token.append('못마땅하다2')
        angry += 1
    if ('정' in comment) & ('떨어지다' in comment):
        check_token.append('정떨어지다2')
        angry += 1
    if ('남' in comment) & ('부끄럽다' in comment):
        check_token.append('남부끄럽다2')
        angry += 1
    # happy
    if ('속' in comment) & ('시원하다' in comment):
        check_token.append('속시원하다')
        happy += 1
    if ('낯' in comment) & ('간지럽다' in comment):
        check_token.append('낯간지럽다')
        happy += 1
    if ('정' in comment) & ('들다' in comment):
        check_token.append('정들다')
        happy += 1
    if ('홀' in comment) & ('가분' in comment):
        check_token.append('홀가분')
        happy += 1

    for token in comment:
        if token in token_happy:
            check_token.append(token)
            happy += 1
        if token in token_sad:
            check_token.append(token)
            sad += 1
        if token in token_disgust:
            check_token.append(token)
            disgust += 1
        if token in token_surprised:
            check_token.append(token)
            surprised += 1
        if token in token_fear:
            check_token.append(token)
            fear += 1
        if token in token_angry:
            check_token.append(token)
            angry += 1
    
    emotion.append(happy)
    emotion.append(sad)
    emotion.append(disgust)
    emotion.append(surprised)
    emotion.append(fear)
    emotion.append(angry)
    check_tokens.append(check_token)
    emotions.append(emotion)
    i += 1
    if i % 400000 == 0:
        print (i)
#     if i > 30 : break
```
