```python
# 데이터 합치기
all_daum_data = pd.concat([df_daum07_ent, df_daum08_ent, df_daum09_ent, df_daum10_ent, df_daum11_ent, df_daum12_ent,
                          df_daum07_non_ent, df_daum08_non_ent, df_daum09_non_ent, df_daum10_non_ent, df_daum11_non_ent, df_daum12_non_ent])
print(all_daum_data.shape)

all_daum_data = all_daum_data.reset_index()
del all_daum_data["index"]
all_daum_data
```

```python
token_data = []
i = 0
for comment in all_daum_data.comment:
    temp = tokenizer_comment(comment)
    token_data.append(temp)
    i += 1                                                                                                 
    
    if i % 100000 == 0:
        print (i)
```
